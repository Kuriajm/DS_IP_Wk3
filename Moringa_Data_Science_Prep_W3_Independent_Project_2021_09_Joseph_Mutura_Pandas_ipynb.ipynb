# 1.2 Merging the dataframes

Merge the dataframes so that we have daily data with their locations together

```python
#Before merging the dataframes lets do some cleaning to remove null rows in all the dataframes

dfd1v1 = dfd1.dropna()
dfd1v1
```

```python
dfd2v1 = dfd2.dropna()
dfd2v1
```

```python
dfd3v1 = dfd3.dropna()
dfd3v1
```

```python
#On geo cell dataset

dfcellsgeov1 = dfcellsgeo.dropna()
dfcellsgeov1
```

This step we now join the three datasets to the cell geo dataset.

```python
#first we rename the columns for consistency

dfd1v1.columns = ['PRODUT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT',
       'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELLID', 'SITE_ID']



#For dataset2
dfd2v1.columns = ['PRODUT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT',
       'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELLID', 'SITE_ID']



#For dataset2
dfd3v1.columns = ['PRODUT', 'VALUE', 'DATETIME', 'CELL_ON_SITE', 'DW_A_NUMBER_INT',
       'DW_B_NUMBER_INT', 'COUNTRY_A', 'COUNTRY_B', 'CELLID', 'SITE_ID']

dfd1v1
```

```python
#Merge starting with the first dataset
#finaldf =pd.merge(combined_dfv4,dfcellsgeo, on= 'CELL_ID')
day1df = dfd1v1.join (dfcellsgeov1)
day1 = day1df.dropna()
```

```python
day2df = dfd2v1.join (dfcellsgeov1)
day2 = day2df.dropna()
day2
```

```python
day3df = dfd3v1.join (dfcellsgeov1)
day3 = day3df.dropna()
day3
```

```python
#We now merge the three days to one dataset for analysis


combined_df = pd.concat([day1,day2,day3])
combined_df
```

#Analysis

```python
#To analyze first we remove the unwanted columns

combined_df = combined_df.drop(columns = ['COUNTRY_A' ,	'COUNTRY_B' ,'CELL_ON_SITE'])
combined_df
```

#Which ones were the most used city for the three days?

```python
result = pd.value_counts(combined_df['VILLES'])
result.head()
```

COCODY is the most used city for the three days with 942 users for the tree days

#Which cities were the most used during business and home hours?

```python
#First we need to convert dates to times we can manipulate

combined_df[['DATE', 'TIME']] = combined_df.DATETIME.str.split(expand=True)
combined_df
```

```python
busycities = combined_df.groupby(['PRODUT', 'VILLES', 'DATE', 'TIME']) ['VALUE'].sum()
busycities
```

```python
busycities = combined_df.groupby(['PRODUT', 'VILLES', 'DATE', 'TIME']) ['VALUE'].sum()
busycities.tail()
```

From this we can say SMS is the least popular service compared to Calls.
And the most active home hours are before midnight.
